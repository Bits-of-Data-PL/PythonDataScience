```python
class Counter:
    def __init__(self, low, high):
        self.current = low - 1
        self.high = high

    def __iter__(self):
        return self

    def __next__(self): # Python 2: def next(self)
        self.current += 1
        if self.current < self.high:
            return self.current
        raise StopIteration


for c in Counter(3, 9):
    print(c)
```

### *Generatory* są mechanizmem
* tworzenia iteratorów
* Zwraca dane przez *yield*
* Każde wywołanie _next()_ zaczyna od miejsca gdzie skończył poprzedni krok
* _next()_ tworzona jest automatycznie


```python
def reverse(data):
    for index in range(len(data)-1, -1, -1):
        print(index)
        yield data[index]
```

```python
for c in reverse('Python'):
    print (c)
    print()
```

```python
mylist = (x*x for x in range(3))
for i in mylist:
    print(i)
```

```python
def create_generator():
    mylist = range(3)
    for i in mylist:
        yield i*i
        
for i in create_generator():
    print(i)
```

```python
def fib(n):
    if n == 0 or n == 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
```

```python
for i in range(36):
    print ("n=%d => %d" % (i, fib(i)))
```

```python
def fib(n):
    a, b = 0, 1
    i=0
    while i < n:
        yield (i, a)
        a, b = b, a + b
        i += 1
```

```python
for i, f in fib(36):
    print ("n=%d => %d" % (i, f))
```
