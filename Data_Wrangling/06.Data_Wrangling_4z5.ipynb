```python
def print_log(message):
    print(message)
    
def noop_log(message):
    pass

def get_log(mode):
    if mode=="dev":
        return print_log
    else: 
        return noop_log
```

```python
env_mode="dev"

log=get_log(env_mode)
```

```python
x=4
log(f'x is {x}')
x*=x
log(f'x is now {x}')
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'value': np.random.randint(0, 100, 20)})
df
```

```python
labels = ["{0} - {1}".format(i, i + 9) for i in range(0, 100, 10)]
```

```python
df['Group'] = pd.cut(df.value, range(0, 105, 10), right=False, labels=labels)
df
```

```python
df.groupby('Group').count()
```

```python
df.groupby('Group').sum()
```

```python
df.groupby('Group').agg({'value': ['count', sum]})
```

```python
def line(series):
    return '#'*len(series)
```

```python
df.groupby('Group').agg({'value': ['count', line]})
```

```python
grouped = df.groupby('Group' ).agg({'value': ['count', line]}).fillna(value=np.nan).replace(to_replace=np.nan, value='')
grouped
```

# Wyrażenia Lambda

```python
def f():
    print ("!")
    
g = f

g()
```

```python
g = lambda: print ("!")

g()
```

```python
a=list(range(-5,0,2))+list(range(0,6,2))
print(a)
```

```python
def square(x):
    return x*x

sorted(a,key = square)
```

```python
sorted(a,key = lambda x: x*x)
```

```python
grouped = df.groupby('Group' ).agg({'value': ['count', lambda x: '#'*len(x)]}).fillna(value=np.nan).replace(to_replace=np.nan, value='')
grouped
```

## Formatowanie

```python
import numpy as np
import pandas as pd

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.randn(8, 3), index=index, columns=['A', 'B', 'C'])
df
```

```python
df.values
```

```python
df.index
```

```python
df.columns
```

```python
newcols = []
for i in range(len(df.columns)):
    newcols.append(df.columns[i].lower())
df.columns = newcols
```

```python
df
```

```python
df.style
```

```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'white'
    return 'color: %s' % color
```

```python
color_negative_red(-1)
```

```python
color_negative_red(2)
```

```python
df.style.applymap(color_negative_red)
```

```python
df.index = pd.to_datetime(df.index, format = '%Y-%m-%d').strftime('%Y-%m-%d')
df.style.applymap(color_negative_red)
```

```python
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: #224477' if v else '' for v in is_max]
```

```python
df.style.apply(highlight_max)
```

```python
df.style.apply(highlight_max, axis=1)
```

```python
df.style.apply(highlight_max, axis=0)
```

```python
df.style.apply(highlight_max, axis=1).applymap(color_negative_red)
```

# `apply`

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(pd.Series(range(10)).values.reshape(5, 2), columns = ['x', 'y'])
df
```

### <span style="color: cyan">APPLY</span>

```python
def ratio(row):
    return row.x/row.y

df['ratio'] = df.apply(ratio, axis=1)
df
```

```python
df['ratio'] = df.apply(lambda row: row.x/row.y, axis=1)
df
```
