### Zadanie 1

Napraw `df_e` biorąc maksimum wartości `baz` dla pary `foo`, `bar` (odrzuć konflikty)

Na postać "tidy" (1 wiersz na ocenę)

```python
import pandas as pd

df_e = pd.DataFrame({"foo": ['one', 'one', 'two', 'two'],
                   "bar": ['A', 'A', 'B', 'C'],
                   "baz": [1, 2, 3, 4]})
df_e
```

```python
df_e.groupby(['foo', 'bar']).aggregate('max')
```

```python
df_e.groupby(['foo', 'bar']).aggregate('max').reset_index()
```

```python
df_e.groupby(['foo', 'bar']).aggregate('max').reset_index().pivot(index='foo', columns='bar', values='baz')  # NO Error
```

### Zadanie 2

Biorąc następujący DataFrame:


```python
df = pd.DataFrame(data = {
    'Day' : ['MON', 'TUE', 'WED', 'THU', 'FRI'], 
    'Google' : [1129,1132,1134,1152,1152], 
    'Apple' : [191,192,190,190,188] 
})
df
```

```python
reshaped_df = df.melt(id_vars=['Day'], var_name='Company', value_name='Closing Price')
reshaped_df
```

Stworzyć korzystając z `reshaped_df` DataFrame identyczny z `df`

```python
original_df = reshaped_df.pivot(index='Day', columns='Company')['Closing Price'].reset_index()
original_df.columns.name = None
original_df
```

```python
reshaped_df.pivot(index='Day', columns='Company')
```

```python
reshaped_df.pivot(index='Day', columns='Company')['Closing Price']
```

```python
original_df = reshaped_df.pivot(index='Day', columns='Company')['Closing Price'].reset_index()
original_df
```

```python
def day_key(day):
    map = {'MON': 0, 'TUE': 1, 'WED': 2, 'THU': 3, 'FRI': 4}
    return map[day]

day_key('TUE')
```

```python
import numpy as np

original_df['day_num'] = np.vectorize(day_key)(original_df['Day'])
```

```python
original_df
```

```python
original_df.sort_values('day_num')
```

```python
original_df.sort_values('day_num').drop('day_num', axis=1).reset_index(drop=True).reindex(columns=['Day', 'Google', 'Apple'])
```
